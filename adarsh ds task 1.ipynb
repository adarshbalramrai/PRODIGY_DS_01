```python
import pandas as pd
import matplotlib.pyplot as plt


sample = [['Uttar Pradesh', 199812341, 16.50, 104480510, 95331831, 9148679, 930, 155111022, 44470455, 240928, 828],
        ['Maharashtra', 112374333, 9.28, 58243056, 54131277, 4111779, 929, 61545441, 50827531, 307713, 365]]


sample = pd.DataFrame(sample, columns=['State/UT', 'Population[50]', 'Percent', 'Male', 'Female', 'Difference between male and female', 'Sex ratio', 'Rural [51]', 'Urban[51]', 'Area[52]', 'Density (per km2)'])

# Create a histogram
plt.figure(figsize=(10, 6))
plt.bar(sample['State/UT'], sample['Male'], label='Male')
plt.bar(sample['State/UT'], sample['Female'], label='Female')
plt.xlabel('State/UT')
plt.ylabel('Population')
plt.title('Population Distribution by Gender in Uttar Pradesh and Maharashtra')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Display the histogram
plt.show()
```
